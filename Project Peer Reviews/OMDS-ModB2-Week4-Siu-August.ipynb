Below are a histogram and boxplot representation of the same data. A pharmacy is keeping a record of the prices of the drugs that it sells, and an administrator wants to know how much the more expensive drugs tend to cost, in the context of the other prices.

Please write a short explanation of the pros and cons of these two representations. Which would you choose? How would you modify the formatting, if at all, to make it more visually interesting, clear, or informative?

```python
import numpy as np
import pandas as pd

np.random.seed(0)
num_data = 100
data = np.exp(np.random.uniform(size = num_data) * 4)
df = pd.DataFrame(data.T, columns = ["data"])
```

```python
print("The 75th percentile is:", df.quantile(q = 0.75))
df.plot.hist()
```

```python
df.plot.box()
```

### Short Explanation of Pros and Cons of Histograms verus Boxplots

### Histogram
#### Pros:
- **Visual Detail**: A histogram shows the distribution of prices across the entire dataset. You can see how many drugs fall into specific price ranges, providing a sense of the overall spread of prices.
- **Frequency Insight**: It allows you to assess where most drugs are priced, helping to visualize the common price range and the density of prices.

#### Cons:
- **Clarity on Extremes**: While it’s easy to see the spread, it can be difficult to pinpoint exact values of individual high-priced drugs. The histogram might obscure the presence of extreme values or outliers if the bin widths are too wide.
- **Less Focus on Specific Values**: The visualization is more about the general trend, not specific points of interest like high prices or outliers.

### Boxplot
#### Pros:
- **Clear Visualization of Extremes**: A boxplot succinctly summarizes the data by showing the median, quartiles, and outliers. This is especially helpful for identifying the "more expensive" drugs, as any data points outside the upper whisker are clearly marked as outliers.
- **Central Tendency**: It gives a quick sense of where most drugs are priced in terms of their median and interquartile range.
- **Outlier Focus**: If the question is about expensive drugs (i.e., outliers), a boxplot makes it very clear where those values fall in the distribution.

#### Cons:
- **Lack of Detail on Distribution**: A boxplot doesn’t provide as much information on the exact distribution, such as how many drugs fall within certain price ranges, and it doesn’t convey the frequency of data points in any given range.
- **Generalization**: While it shows where most of the data lies, it doesn’t provide as much insight into variations within each quartile.

### Which one is better to use for dataset representation?
- If the primary concern is understanding how much the more expensive drugs tend to cost in relation to other prices, the **boxplot** is probably the better option because it highlights the high-end outliers clearly. It directly addresses the question of expensive drugs by showing them as outliers above the upper whisker.
- If the administrator is also interested in the frequency and spread of prices (i.e., how many drugs are near a specific price), a **histogram** would be more suitable. 

### Modifications for Improved Visual Appeal and Clarity:
- **Histogram**: 
  - **Bin Size**: Ensure the bin size is not too large or too small—this could make the distribution too generalized or too noisy.
  - **Color Coding**: Use color to highlight important sections, like marking the high-end bins with a contrasting color to emphasize the more expensive drugs.
  - **Overlaying a Boxplot**: If you want the best of both worlds, you could overlay a boxplot onto the histogram to show the distribution and key summary statistics simultaneously.

- **Boxplot**:
  - **Notched Boxplot**: You might consider using a notched boxplot to provide more insight into the confidence intervals of the median.
  - **Color**: Use color to highlight outliers or the quartiles, so they stand out visually, especially if there are multiple categories or subgroups (like different types of drugs).
  - **Labeling**: Clearly label the whiskers and any outliers, so there’s no ambiguity about what’s being represented.

For answering the specific question of "how much the more expensive drugs tend to cost," I would recommend the **boxplot** for its clarity in showing outliers. However, a combination of both representations—perhaps with the histogram for the overall distribution and a boxplot for outliers—might provide a fuller picture.

```python
import pandas as pd
import os

# Set the path to the downloaded dataset
# Make sure to adjust the path to your local Kaggle folder and the dataset file name
file_path = os.path.join('/Users/augustasfour/downloads', 'US_Accidents_March23.csv')  # Change this to your actual file path

# Load the dataset into a DataFrame
df_dataset2 = pd.read_csv(file_path)

# Display the first few rows
print(df_dataset2.head())
```

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python

continuous_columns = ['Distance(mi)', 'Start_Lat', 'Start_Lng', 'End_Lat', 'End_Lng']  # Add more columns if needed

plt.figure(figsize=(15, 10))
for i, column in enumerate(continuous_columns, 1):
    plt.subplot(2, 3, i)
    df_dataset2[column].hist(bins=20, edgecolor='black', alpha=0.7)
    plt.title(f'Histogram of {column}')
    plt.xlabel(column)
    plt.ylabel('Frequency')

plt.tight_layout()
plt.show()
```

```python
plt.figure(figsize=(10, 6))
sns.histplot(data=df_dataset2, x='Distance(mi)', hue='Severity', kde=True, bins=10, multiple='stack')
plt.title('Grouped Histogram of Distance by Severity')
plt.xlabel('Distance')
plt.ylabel('Frequency')
plt.show()
```

```python

plt.figure(figsize=(10, 6))
sns.countplot(x='Severity', data=df_dataset2, palette='Set2')
plt.title('Bar Plot of Severity Counts')
plt.xlabel('Severity')
plt.ylabel('Count')
plt.show()

plt.figure(figsize=(10, 6))
sns.countplot(x='Source', data=df_dataset2, palette='Set2')
plt.title('Bar Plot of Source Counts')
plt.xlabel('Source')
plt.ylabel('Count')
plt.xticks(rotation=45)
plt.show()
```

```python

continuous_stats = df_dataset2[continuous_columns].agg(['mean', 'median', lambda x: x.mode()[0]])
print(continuous_stats)

```
